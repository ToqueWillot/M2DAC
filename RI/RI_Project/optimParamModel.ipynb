## Optimisation du parametre k du modèle OKAPI

#### Récupération des scores

```python
fichier=open("okapiOpti.csv")

k=[]
score=[]
for line in fichier.readlines():
    for tup in line.split(";")[:-1]:
        k1=float(tup.split(",")[0])
        s1=float(tup.split(",")[1])
        k.append(k1)
        score.append(s1)
```

#### Affichage des scores en fonction du paramètre

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.plot(k,score)
plt.xlabel("k")
plt.ylabel("Score")
plt.title("Precision moyenne du modele OKAPI en fonction du parametre k")
plt.show()
plt.close()
```
